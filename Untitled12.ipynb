```python
import re
import string
from collections import Counter
from typing import List,Tuple,Dict
```

```python
def preprocess_text(text):

  text=text.lower()

  text=re.sub(r'http\S+|www\S+','',text)

  text=re.sub(r'\S+@\S+','',text)

  text=re.sub(r'\d+','',text)

  text=re.sub(r'(.)\1{2,}',r'\1',text)

  text=re.sub(r'[^\w\s]','',text)

  text=re.sub(r'\s+',' ',text).strip()

  words=text.split()
  filtered_words=[]

  for word in words:
      if len(word)>2 or word in ['no','not']:
        filtered_words.append(word)


  return ' '.join(filtered_words)


```

```python
text="I have 2 dogs WOW!!! This is GREAT!!! Visit http://example.com now "
print(preprocess_text(text))
```

```python
test_sentences=[
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
for sentence in test_sentences:
  cleaned=preprocess_text(sentence)
  tokens=cleaned.split()

  total_tokens=len(tokens)
  unique_tokens=len(set(tokens))

  if total_tokens >0:
    avg_length=sum(len(word) for word in tokens) / total_tokens
  else:
    avg_length=0

  print("Original Text:",sentence)
  print("Cleaned Tokens:",tokens)
  print("Cleaned sentence:",cleaned)
  print("Total tokens:",total_tokens)
  print("Unique tokens:",unique_tokens)
  print("Avg token length:",round(avg_length,2))
  print("-"* 50)
```

```python
from collections import Counter

all_tokens=[]
for sentence in test_sentences:
  cleaned=preprocess_text(sentence)
  tokens=cleaned.split()
  all_tokens.extend(tokens)

  word_freq=Counter(all_tokens)

  top_10=word_freq.most_common(10)

  print("top 10 most frequent words:")
  for word, count in top_10:
    print(word, ":", count)

```

```python
least_5=word_freq.most_common()[-5:]
print("\nTop 5 least frequent words!")
for word,count in least_5:
    print(word,":",count)
```

```python
def full_pipeline(text_list):
  all_tokens=[]
  clean_sentences=[]
  for text in text_list:
    if not isinstance(text,str)or text.strip()=="":
      clean_sentences.append("")
      continue
    cleaned=preprocess_text(text)

    if cleaned=="":
      clean_sentences.append("")
      continue

    tokens=cleaned.split()
    all_tokens.extend(tokens)
    clean_sentences.append(cleaned)
  return{
    "tokens":all_tokens,
    "clean_sentences":clean_sentences
  }
```

```python
test_data=[
    "hello Innov!!!",
    "",
    "😍😍😍",
    "1234568",
    "Visit http://example.com now"

]

result=full_pipeline(test_data)
print(result)
```
